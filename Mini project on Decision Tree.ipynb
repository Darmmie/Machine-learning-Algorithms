```python
import pandas as pd
```

```python
#upload working data and name it salary

salary= pd.read_csv("salaries.csv")
salary
```

```python
#Exploratory Data Analysis (EDA)
#viewing the first five rows of our data to help us understand it for the analysis
salary.head()
```

```python

# From the above table, we can see that salary is a dependent variable, while other variable are independent.
# Name a new table (input) to the main table where "salary" a dependent variable has been dropped.
input = salary.drop("salary_more_then_100k", axis = "columns")
input
```

```python
# Make the dependent variable a table, named "target"

target=salary["salary_more_then_100k"]
target
```

```python
# converting input data to dummy variables to numerical categories 
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode the variable company, job and degree by using LabelEncoder

le_company = LabelEncoder()
le_job= LabelEncoder()
le_degree = LabelEncoder()
```

```python
# Use fit_transform to transform "le_" in the input table.
input["company_x"] = le_company.fit_transform(input["company"])
input["job_x"] = le_job.fit_transform(input["job"])
input["degree_x"] = le_degree.fit_transform(input["degree"])
```

```python
# Call the table to see if all variables have transformed.

input
```

```python
# Drop the initial Columns Variables(company, job, degree), and name the the table final_dummy

final_dummy =input.drop(["company", "job", "degree"],axis = "columns")
final_dummy 
```

```python
target
```

```python
# Import decision tree model from sklearn
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
# Model fit our dependent and independent variables

model.fit(final_dummy, target)
```

```python
# find the model score. in this table the model score is overfitting because we use all of our data for training.
model.score(final_dummy, target)
```

### To answer the question:
Is salary of a google computer programmer with a bachelor degree > 100k?

```python
# Use our model to predict. google = 2, comp.prog = 1, and bachelors deg = 0

model.predict ([[2,1,0]])
```

 0 output informs us a computer programmer with a bachelors degree will not earn above 100k
